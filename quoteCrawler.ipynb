```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
from collections import defaultdict
```

```python
def firstCrawl(url):
  page = requests.get(url)
  soup = BeautifulSoup(page.text,'html.parser')
  f = open('kq.txt', 'wb')
  f.write(soup.encode('utf-8'))
  f.close()
firstCrawl('http://quotes.toscrape.com/')
```

```python
quotes = defaultdict(list)
result = []
pages_crawled = []
authors = defaultdict(list)
author_url = []
```

```python

def siteCrawler(url):
    
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    links = soup.find_all('a')
    quotediv = soup.find_all('div', class_='quote')
    result.append(quotediv)
    for element in quotediv:
        author = element.find('small', class_='author')
        quote = element.find('span', class_='text').text
        if quote not in quotes:
          quotes[author.text] += [quote]
        print(author)
    #1.2b
    for link in links:
        if 'href' in link.attrs:
            
            if link['href'].startswith('/') and ':' not in link['href']:
                if link['href'].startswith('/author'):
                    if (f"http://quotes.toscrape.com{link['href']}") not in author_url:
                        author_url.append(f"http://quotes.toscrape.com{link['href']}")
                if link['href'] not in pages_crawled:
                    new_link = f"http://quotes.toscrape.com{link['href']}"
                    pages_crawled.append(link['href'])
                    siteCrawler(new_link)
siteCrawler('http://quotes.toscrape.com/')
```

```python
#1.2a
print(result)
```

```python
#1.2c
authorInfo = []
def tacgiaLink(author_url):
  for url in author_url:
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    author_name = soup.find('h3', class_='author-title').text
    author_born_date = soup.find('span', class_='author-born-date').text
    print(author_name)
    print(url)
    print(author_born_date)
    print(quotes[author_name])
    print('=====================================NEXT=====================================')
    authorInfo.append((author_name,url,author_born_date,quotes[author_name]))
tacgiaLink(author_url)
```

```python
#1.2d
data = [[a for a in item] for item in authorInfo]
df_data = pd.DataFrame(data,columns = ['Author','Author URL','DOB','Quotes'])
df_data.to_csv('Quote.csv')
df_data.head(15)
```
